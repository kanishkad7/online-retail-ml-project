```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/kaggle/input/datasets/mashlyn/online-retail-ii-uci/online_retail_II.csv")

# Clean data
df.dropna(inplace=True)

# Create Total Price
df['TotalPrice'] = df['Quantity'] * df['Price']

df.head()
```

```python
plt.scatter(df['Quantity'], df['TotalPrice'])
plt.title("Quantity vs Revenue")

plt.savefig("/kaggle/working/RQ1_figure.pdf")

df[['Quantity','TotalPrice']].to_csv("/kaggle/working/RQ1_table.csv", index=False)
```
